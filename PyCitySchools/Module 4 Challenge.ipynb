```python
import os
import csv
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="outer", on=["school_name","school_name"])
school_data_df = pd.DataFrame(school_data_complete)

grouped_school = school_data_df.groupby(["school_name"])
grouped_school.count()
```

```python

name_type_df = school_data[["school_name","type"]]
name_budget_df = school_data[["school_name","budget"]]
```

```python

#number student that passed math, reading and overall
grouped_school_math = school_data_df[school_data_df["math_score"] >= 70].groupby(["school_name"]).count()
grouped_school_reading = school_data_df[school_data_df["reading_score"] >= 70].groupby(["school_name"]).count()
grouped_school_overall = school_data_df[(school_data_df["math_score"] >= 70) & (school_data_df["reading_score"] >= 70)].groupby(["school_name"]).count()

#extracting the two series from the dataframe and merging to create a mini dataframe:
passed_math_df = grouped_school_math[["math_score"]]
passed_reading_df = grouped_school_reading[["reading_score"]]
math_reading_df = pd.merge(passed_math_df, passed_reading_df, how="outer", on=["school_name"])

overall_df = grouped_school_overall[["Student ID"]]

# merging the math and reading dataframe with the overall passing series:
three_pass = pd.merge(math_reading_df, overall_df, how="outer", on=["school_name"])
three_pass_rename = three_pass.rename(columns={'math_score':'# of Student Passed Math',
                                                'reading_score': '# of Student Passed Reading',
                                               'Student ID':'# of Student Passed Both'})
three_pass_rename
```

```python
# total schools:
total_schools = school_data_df["school_name"].nunique()
# total students:
total_students = school_data_df["student_name"].count()
# total budget:
schools_budget_df = school_data_df.filter(['school_name','budget'], axis = 1)
unique_shcools_budget_df = schools_budget_df["budget"].unique()
total_budget = unique_shcools_budget_df.sum()
# average math score:
average_math_score = school_data_df["math_score"].mean()
# average reading score:
average_reading_score = school_data_df["reading_score"].mean()
# %passing math:
math_passed = school_data_df.loc[school_data_df['math_score'] >= 70]
math_students = math_passed['math_score'].count()
math_pct = math_students/total_students
# %passing reading:
read_passed = school_data_df.loc[school_data_df['reading_score'] >= 70]
read_students = read_passed['reading_score'].count()
read_pct = read_students/total_students
# %passing overall:
overall_pass_students = school_data_df.loc[(school_data_df['math_score'] >= 70) & (school_data_df['reading_score'] >= 70)].count()
overall_pct = overall_pass_students/total_students
overall_pct["Student ID"]

District_summary_df = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students],
                                   "Total Budget": [total_budget], "Average Math Score": [average_math_score],
                                   "Average Reading Score": [average_reading_score],"% Passing Math": [math_pct],
                                   "% Passing Reading": [read_pct], "% Overall Passing": [overall_pct["Student ID"]]})

District_summary_df
```

```python
clean_data = school_data_df.drop(columns = ['student_name','Student ID','gender','grade','School ID'])
grouped_byschool = clean_data.groupby(["school_name"])
average_math_reading = grouped_byschool[['reading_score','math_score']].mean()
grouped_byschool_df = grouped_byschool.count()

school_summary_df = pd.merge(grouped_byschool_df, average_math_reading, how="outer", on=["school_name"])
school_summary_df

type_merge = pd.merge(school_summary_df, name_type_df, how="outer", on=["school_name"])
budget_merge = pd.merge(type_merge, name_budget_df, how="outer", on=["school_name"])

# Adding a new column that calculates per student budget:
budget_merge["Per Student Budget"] = (budget_merge["budget_y"] / budget_merge["size"])

# Merge dataframe with the dataframe that has number of student passing & droping a few more columns with no meaning:
school_summary_1 = pd.merge(budget_merge, three_pass_rename, how="outer", on=["school_name"])
school_summary_2 = school_summary_1.drop(columns = ['reading_score_x','math_score_x','type_x','budget_x'])

# Adding columns that calculates the passing percentage for math, reading and overall:
school_summary_2["% Passing Math"] = (school_summary_2["# of Student Passed Math"] / school_summary_2["size"])
school_summary_2["% Passing Reading"] = (school_summary_2["# of Student Passed Reading"] / school_summary_2["size"])
school_summary_2["Overall Passing"] = (school_summary_2["# of Student Passed Both"] / school_summary_2["size"])

# Droping the student number columns:
school_summary_3 = school_summary_2.drop(columns = ['# of Student Passed Math','# of Student Passed Reading','# of Student Passed Both'])
school_summary_3

# Renaming and organizing the colums:
school_summary_4 = school_summary_3.rename(columns={'size':'Total Students',
                                                    'reading_score_y': 'Average Reading Score',
                                                    'math_score_y': 'Average Math Score',
                                                    'type_y':'School Type',
                                                    'budget_y':'Total School Budget',})
School_Summary_Final = school_summary_4[['school_name','School Type','Total Students','Total School Budget','Per Student Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','Overall Passing']]

School_Summary_Final
```

```python
school_summary_sort = School_Summary_Final.sort_values('Overall Passing', ascending=False)
school_summary_sort.head(5)
```

```python
school_summary_sort = School_Summary_Final.sort_values('Overall Passing')
school_summary_sort.head(5)
```

```python
studentdata_df = pd.DataFrame(student_data)
math_by_grade = studentdata_df.pivot_table(values='math_score', index ='school_name', columns = 'grade', aggfunc = 'mean')
math_by_grade
```

```python
reading_by_grade = studentdata_df.pivot_table(values='reading_score', index ='school_name', columns = 'grade', aggfunc = 'mean')
reading_by_grade
```

```python

# Creating bins and group names, creating new column and groupby:
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
School_Summary_Final = School_Summary_Final.copy()
School_Summary_Final["Spending Ranges"] = pd.cut(School_Summary_Final["Per Student Budget"], bins, labels=group_names, include_lowest=True)
Grouped_by_budget = School_Summary_Final.groupby("Spending Ranges").mean(numeric_only=True)


# Droping columns that are not required:
Grouped_by_budget.drop(columns = ['Total Students','Total School Budget','Per Student Budget'], inplace=True)
```

```python
bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
School_Summary_Final["School Size"] = pd.cut(School_Summary_Final["Total Students"], bins, labels = group_names, include_lowest = True)
Grouped_by_size = School_Summary_Final.groupby("School Size").mean()

# Droping columns that are not required:
Grouped_by_size.drop(columns = ['Total Students','Total School Budget','Per Student Budget'])
```

```python
Grouped_by_type = School_Summary_Final.groupby("School Type").mean()
Grouped_by_type.drop(columns = ['Total Students','Total School Budget','Per Student Budget'])
```
